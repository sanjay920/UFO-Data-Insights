```python
import datapackage
import pandas as pd
```

```python
data_url = 'https://datahub.io/core/airport-codes/datapackage.json'
package = datapackage.Package(data_url)
resources = package.resources
for resource in resources:
    if resource.tabular:
        airport_df = pd.read_csv(resource.descriptor['path'])
        airport_df = airport_df[airport_df.iso_country == "US"]
```

```python
airport_df
```

```python
df = airport_df[((airport_df.type == 'large_airport') == True) | ((airport_df.type == 'medium_airport') == True) | ((airport_df.type == 'small_airport') == True)]
```

```python
df[['id', 'type', 'latitude_deg', 'longitude_deg']].to_csv('airports.csv', index=False)
```
